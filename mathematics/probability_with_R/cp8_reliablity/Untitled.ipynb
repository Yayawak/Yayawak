# Rel of **Parallel System**
## $$
\begin{equation}
\begin{split}
R &= P(\text{at least one of the components is working}) \\
    &= 1 - P(\text{all components fail}) \\
    &= 1 - P(\text{1st fails}) P(\text{2st fails}) \dots P(\text{nth fails}) \\
    &= 1 - (1-p_1)(1-p_2) \dots (1-p_i) \\
    &= 1 - \prod_{i=1}^{n} (1-p_i)
\end{split}
\end{equation}
$$

```r
# Ex 8.4
p_each_working = .97
Rel = 1 - (1-.97)^5
Rel
```

```r
k = 1:10
p = .97
plot(k, 1-(1-p)^k,
     xlab="Number of components",
     ylab="Rel"
    )
abline(1,0)
```
