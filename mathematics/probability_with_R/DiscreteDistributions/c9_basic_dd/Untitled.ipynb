# Cumulative Distribution Function (CDF)
## $$
\begin{equation}
\begin{split}
E = \{X \leq 2 \} &= \{X=0\} + \{X=1\} + \{X=2\} \\
\\
    P(X \leq 2) &= P(X=0) + P(X=1) + P(X=2)
\end{split}
\end{equation}
$$

# $$
F(x) = \sum_{k\leq x}p(k)
