# P(all jobs are assigned to different processors)

```r
# n processors
n_procs = 400
n_jobs = 10
prod( n_procs : (n_procs-(n_jobs-1)) ) / n_procs^n_jobs
# prod(400: (400-9)) / 400^10
```

```r
n_job = 10
n_proccessor = 600

diffprocessor = 0
for (n in (n_job: n_proccessor)) {
# for (n in 10: 400) {
    diffprocessor[n] = prod(n: (n-(n_job-1)) ) / (n^10) 
}

k = seq(n_jobs, n_proccessor, 10)
plot(k, diffprocessor[k],
    xlab="Number of processors available",
    ylab="P(no processors recieves more than one job)")

abline(h=.9)
```
