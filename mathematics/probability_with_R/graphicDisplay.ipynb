```r
females_score = c(sample(40:100, 60))
males_score = c(sample(40:100, 40)) * 1.1
```

```r
gender_marks = rep(c("W", "M"), c(60, 40))
gender_marks
```

```r
# females_score + males_score
mixed_scores = c(females_score, males_score)
mixed_scores
```

```r
mixed_scores~gender_marks
```

```r
boxplot(mixed_scores~gender_marks,
       xlab='Gender', ylab="Score", 
       main="Boxplot")
```

```r
par(mfrow=c(2,2))
boxplot(mixed_scores~gender_marks)
```
